8. SELECT
  - 1) 기본 문법
    - SELECT 필드명, 필드명, ...FROM 테이블명;

           SELECT ename, sal, deptno, job FROM emp;

    - 모든 필드 불러오기
           SELECT *FROM 테이블명;

    - 필드의 가공처리(+ 숫자)
           SELECT empno, ename, sal, sal+200 FROM emp;(sal에 200을 더해준 값을 표시해줌)
 
    - 필드의 alias(as를 사용하여 별명 설정)
           SELECT ename, mgr as manager, comm as bonus FROM emp;

    - 중복 제거(distinct)
           SELECT distinct job FROM emp;

  - 2) 확장 문법
    - 정렬 기능 : ORDER BY 필드명[ASC | DESC], 필드명, ... (ASC 오름차순, DESC 내림차순, 그냥 생략하면 오름차순)
      - 급여가 많은 순으로 이름, 급여, 부서코드 조회
             SELECT ename, sal, deptno FROM emp ORDER BY sal DESC;

      - 부서별로 급여가 적은 순으로 이름, 급여, 부서코드 조회
             SELECT ename, sal, deptno FROM emp ORDER BY deptno, sal DESC;

    - 조건 기능(레코드 필터링) : WHERE 조건문
      - 급여가 3000$ 이상인 직원의 사번, 이름, 업무, 급여를 조회
             SELECT empno, ename, job, sal FROM emp WHERE sal >= 3000;
 
      - 업무가 manager인 직원의 이름, 부서, 업무, 급여를 조회
             SELECT ename, deptno, job, sal FROM emp WHERE job = 'manager'; 
        (SQL은 index가 0이 아닌 1부터 시작)
      
      - 부서가 20이고 업무가 analyst인 직원의 이름, 부서, 업무, 급여를 조회
             SELECT ename, deptno, job, sal FROM emp WHERE deptno = 20 AND job = 'analyst';

      - 급여가 1500이상 2500이하를 받는 직원의 이름, 부서, 업무, 급여를 조회
             SELECT ename, deptno, job, sal FROM emp WHERE sal >= 1500 AND sal <= 2500;
             SELECT ename, deptno, job, sal FROM emp WHERE sal BETWEEN 1500 AND 2500;

      - 급여가 1500이상 2500이하를 받는 직원의 이름, 부서, 업무, 급여를 조회(단, 급여가 많은 순으로 조회)
             SELECT ename, deptno, job, sal FROM emp WHERE sal BETWEEN 1500 AND 2500 ORDER BY sal DESC; (ORDER BY는 항상 마지막)

      - 업무가 clerk, salesman, analyst인 직원의 이름, 부서, 업무, 급여를 조회 
             SELECT ename, deptno, job, sal FROM emp WHERE job = 'clerk' OR job = 'salesman' OR job = 'analyst';
             SELECT ename, deptno, job, sal FROM emp WHERE job IN ('clerk', 'salesman', 'analyst');

      - 1982년 1월 1일 이후에 입사한 직원의 이름, 업무, 입사일자를 조회
             SELECT ename, job, HIREDATE FROM emp WHERE HIREDATE > '1982-01-01';
             (sql에서는 날짜와 문자열이 자동 변환된다)

      - 커미션이 없는 직원의 이름, 부서, 업무, 급여, 커미션을 조회
             SELECT ename, deptno, job, sal, comm FROM emp WHERE comm is Null OR comm = 0;
             SELECT ename, deptno, job, sal, comm FROM emp WHERE job != 'salesman';
             NULL 은 = 이 아닌 is로 같음을 표현한다
      
      - 커미션이 있는 직원의 이름, 부서, 업무, 급여, 커미션을 조회
             SELECT ename, deptno, job, sal, comm FROM emp WHERE comm is NOT NULL;


  - 3) Like 연산자
    - % : 0개 이상의 문자를 대체
    - _ : 1개 문자를 대체<br>
      beau% : beau, beau로 시작하는 모든 문자열
      beau_ : beaua, beaub, ..
    - 이름이 s로 시작하는 직원의 이름, 업무, 급여를 조회
      SELECT ename, job, sal FROM emp WHERE ename LIKE 's%';

  - 4) GROUP BY 필드명[, 필드명, 필드명, ...][HAVING 조건문] (HAVING은 무조건 GROUP BY에서만)
    - 부서별로 급여 평균, 최저 급여를 조회
          SELECT deptno, avg(sal), min(sal) FROM emp GROUP BY deptno;

    - 부서별로 급여 평균, 최저 급여를 조회(단, 급여 평균이 높은 순으로 조회)
          SELECT deptno, avg(sal) AS '급여평균', min(sal) AS '최저급여' FROM emp GROUP BY deptno ORDER BY avg(sal) DESC;

    - 전체 급여의 합계가 5000을 초과하는 업무에 대해 급여 합계를 조회
          SELECT job, SUM(sal) FROM emp GROUP BY job having sum(sal) > 5000;
 
    - 전체 급여의 합계가 5000을 초과하는 업무에 대해 급여 합계를 조회(단, salesman은 제외)
          SELECT job, SUM(sal) FROM emp GROUP BY job HAVING sum(sal) > 5000 AND job != 'salesman';
          SELECT job, SUM(sal) FROM emp WHERE job != 'salesman' GROUP BY job HAVING sum(sal) > 5000;

9. 함수 (Single Row Function, Aggregation Function, ...)
  - 1) Single Row Function(String, Nymeric, Data&Time)
      - String Function
            - ASCII(), CHAR()
              SELECT ascii('a'), char(98);

            - LENGTH()
              SELECT ename, length(ename) FROM emp;

            - INSTR()
              SELECT instr('foobarbar', 'bar');, SELECT instr('foobarbar', 'love');
              
            - LEFT(), RIGHT(), SUBSTR(), SUBSTRING()
              SELECT left('abcdefg', 3);
              SELECT right('abcdefg', 3);
              SELECT substring('l love sql', 3); (시작위치[3]부터 끝까지)
              SELECT substring('l love sql', 3, 4); (시작위치[3]부터 4글자)
              SELECT substring('l love sql', -3);
              SELECT substring('l love sql', -8, 3);

            - LOWER(), UPPER()
              SELECT lower('ABcdEF'), upper('ABcdEF');

            - LTRIM(), RTRIM(), TRIM()
              SELECT LTRIM('      abc      ') as trim;
              SELECT RTRIM(LTRIM('      abc      ')) as trim;
              SELECT TRIM('      abc      ') as trim;

      - Numeric Function

            - CEIL(), FLOOR(), ROUND(), TRUNCATE() - 무조건 버림
              SELECT ceil(1.23), floor(1.78);
              SELECT round(1.23), round(1.23, 0);
              SELECT round(1.78, 1); (소수점 첫번째자리에서 반올림)
              SELECT truncate(1.78, 0), truncate(1.78, 1), truncate(1.78, -1); (몇번째 자리에서 무조건 버림)

            - MOD() - 나머지
              SELECT mod(20, 9), 20 mod 9, 20%9;

            - RAND() - 0부터 1사이의 난수 뽑기
              SELECT rand();
              SELECT floor(rand()*(m-n)+n);

              7보다 크거나 같고 12보다 작은 범위내에서 난수를 뽑아라.
              SELECT floor(rand()*(12-7)+7);

      - Date&Time Function
            - NOW(), SYSDATE(), CURDATE(), CURTIME()
              SELECT now();
              SELECT sysdate();
              SELECT curdate();
              SELECT curtime();

            - YEAR(), MONTH(), DAYOFMONTH() : 연, 월 ,일
              SELECT year(now()), month(now()), dayofmonth(now());
              SELECT dayofmonth('2021/05/19');

            - DAYOFWEEK() : 1-일요일, WEEKDAY() : 0-월요일
              SELECT dayofweek(now()), weekday(now());

            - DAYOFYEAR() : 현재까지의 일수
              SELECT dayofyear(now());

            - DATE_ADD(), DATE_SUB(), ADDDATE()
              SELECT DATE_ADD(now(), interval 5 year);
              SELECT DATE_ADD(now(), interval 100 day);

            - 그외
              SELECT datediff(now(), '2019/05/19');
              SELECT last_day(now()); -> 달의 마지막 날짜를 구해준다

  - What?
      - 실체(Entity) 무결성 : 중복된 데이터를 방지
            - Primary Key, Unique(제약 = Constraint)
      - 영역(Domain) 무결성 : 범위
            - Check
      - 참조(Reference) 무결성
            - Foreign Key

  - 컬럼의 속성
      - NN(Not NULL -> 절대 값을 비워 놓을 수 없다) 속성
            - Primary Key
            - Not NULL(테이블을 만들때 설정해준다)
      - ND(No Duplicate -> 같은 값이 있어서는 안된다) 속성
            - Primary Key
            - Unique
      - NC(No Change -> 값이 변할 수 없다) 속성
            - Foreign Key

  - Primary Key 특성
          - NN속성과 ND속성 지원
          - 한 테이블에 한 개만 사용
          - 여러 개의 컬럼을 묶어서 사용 가능

  - 연습용 테이블

          CREATE TABLE tblExam(
            id    int,
            name  varChar(10)
          );
        
          (varChar은 가변길이, char은 고정길이)   

  - *** Primary Key ***
      - 기본키 없을때
            INSERT INTO tblExam(id) VALUES(1);
            INSERT INTO tblExam(name) VALUES('hong');
            --> NULL 값이 생성된다
            DELETE FROM tblExam; 으로 싹 지워준다

      - ALTER TABLE을 사용한다(ADD, DROP, MODIFY)

      - 기본키 추가          
            ALTER TABLE tblExam ADD CONSTRAINT pk_id PRIMARY KEY(id);
            --> id에 기본키 추가

      - NN속성 확인    
            INSERT INTO tblExam(name) VALUES('hong');
            --> id값이 NULL이기 때문에 값이 생성되지 않는다

      - ND속성 확인
            INSERT INTO tblExam(id) VALUES(1);
            INSERT INTO tblExam(id) VALUES(1);
            --> id값이 중복 생성될 경우 값이 생성되지 않는다

      - 기본키 삭제
            ALTER TABLE tblExam DROP PRIMARY KEY;

      - 기본키 삭제 후 ND속성 확인
            INSERT INTO tblExam(id) VALUES(1);
            --> 키가 삭제되었기 떄문에 중복된 값이 생성된다

      - 테이블 작성시 기본키 추가

            CREATE TABLE tblExam(
              id    int    PRIMARY KEY,
              name  varChar(10)
            );
            --> 하나의 컬럼에만 키를 설정한다

            또는

            CREATE TABLE tblExam(
              id    int,
              name  varChar(10),
              CONSTRAINT pk_id PRIMARY KEY(id, name)
            );
            --> 중복된 컬럼에도 키를 설정할 수 있다

      - NN, ND 속성 확인
            INSERT INTO tblExam VALUES(1, 'hong');
            INSERT INTO tblExam VALUES(1, 'kim');
            INSERT INTO tblExam VALUES(1, 'hong');
            --> 세번째는 중복(ND)이기에 추가되지 않는다

  - 참고
      - tel 필드 추가
            ALTER TABLE tblExam ADD tel char(10) not null;

      - tel 필드를 NULL로 변경 --> MODIFY 활용
            ALTER TABLE tblExam MODIFY tel char(15) null;

      - tel 필드 삭제
            ALTER TABLE tblExam DROP tel;

  - *** Unique ***
      - 테이블 작성시 유니크 키 추가
            CREATE TABLE tblExam(
              id    int    Unique,
              name  varChar(10) Unique
            );

            또는

            CREATE TABLE tblExam(
              id    int,
              name  varChar(10),
              CONSTRANINT uk_id UNIQUE(id),
              CONSTRANINT uk_name UNIQUE(name)
            );

      - ND 속성 지원 여부
            INSERT INTO tblExam(id) VALUES(1);
            INSERT INTO tblExam(id) VALUES(1);
            --> ND 지원

      - NN 속성 테스트
            INSERT INTO tblExam(name) VALUES(null);
            INSERT INTO tblExam(name) VALUES(null);
            --> NN 지원 X

  - NN과 ND를 막아주는 또 다른 장치
      - NN -> default : 값이 주어지지 않을 때 디폴트 값을 넣어준다
      
      - ND -> auto_increment : 자동으로 순차적인 값이 생성된다(단, 내가 값을 입력할 수 없다)
      
      - Default와 auto_increment가 있는 테이블 생성
            CREATE TABLE tblExam(
              id    int          auto_increment,
              name  varchar(10)  default 'no name',
              age   int          default 0,
              CONSTRAINT uk_id UNIQUE(id)
            );
      
      - Default와 auto_increment가 잘 실행되는지 확인
            INSERT INTO tblexam(NAME) VALUES('hong');
            --> age가 디폴트 값인 0으로 설정

            INSERT INTO tblexam(age) VALUES(25);
            --> name이 디폴트 값인 no name으로 설정

            --> 둘다 id가 순차적으로 증가

  - *** Foreign Key ***
      - 부모 테이블, 자식 테이블은 참조관계
      - 자식 테이블에 Foreign Key(참조키)를 달아준다
      - Dept 테이블 작성
            CREATE TABLE tblDept(
              deptno    int,
              dname     varchar(10)
            );

            INSERT INTO tblDEPT VALUES(100, 'Sales');
            INSERT INTO tblDEPT VALUES(200, 'Research');
      
      - Emp 테이블 작성
            CREATE TABLE tblEmp(
              empno     int,
              ename     varchar(10),
              hiredate  datetime,
              deptno    int
            );

            INSERT INTO tblEmp(empno, ename, hiredate, deptno) VALUES(1, 'Hong', now(), 300);
            --> deptno에 해당하지 않는 300도 별다른 제재 없이 추가가 가능하다

      - Foreign Key 추가
            ALTER TABLE tblEmp ADD CONSTRAINT fk_deptno FOREIGN KEY(deptno) REFERENCES tblDept(deptno);
            --> 에러가 난다. 부모 테이블에 PrimaryKey나 Unique가 없기 때문
            
      - 부모 테이블(tbldept)의 deptno에 PrimaryKey 설정해준 다음 Foreign Key 추가
            ALTER TABLE tbldept ADD CONSTRAINT pk_deptno PRIMARY KEY(deptno);
            ALTER TABLE tblEmp ADD CONSTRAINT fk_deptno FOREIGN KEY(deptno) REFERENCES tblDept(deptno);

      - Foreign Key가 잘 작동되는지 확인
            INSERT INTO tblEmp(empno, ename, hiredate, deptno) VALUES(1, 'Hong', now(), 300);
            --> deptno가 300으로 설정되어 있기 때문에 추가되지 않는다
            INSERT INTO tblEmp(empno, ename, hiredate, deptno) VALUES(1, 'Hong', now(), 200);
            --> 정상적으로 추가

      - Foreign Key의 NC 속성이 잘 작동되는지 확인
            UPDATE tblDept
            SET deptno = 2000
            WHERE deptno = 200;
            --> Foreign Key 때문에 변경되지 않는다

            UPDATE tblDept
            SET deptno = 1000
            WHERE deptno = 100;
            --> dept = 100인 값이 존재하지 않기 때문에 실행자체는 된다

          --> Possible_keys가 NULL

16. 트랜젝션(Transaction)
    - 특징 -> ACID
          Atomicity    원자성
          Consistency  일관성
          Isolation    고립성
          Durability   지속성
    
    - 임의로 정해 놓은 작업(DML) 단위
    - "ALL or NOTHING"

    - rollback
          DELETE FROM tblindex;
          SELECT * FROM tblindex;
          --> 데이터가 전부 삭제되어 있다
          
          ROLLBACK;
          SELECT * FROM tblindex;
          --> 데이터가 다시 복구

          DELETE FROM tblindex;
          COMMIT;
          ROLLBACK;
          SELECT * FROM tblindex;
          --> 데이터가 복구되지 않는다

    - commit(commit이 되면 rollback을 할 수 없다)
      - commit을 수동으로 바꿔주기
            SELECT @@autocommit;
            --> 1(자동)
            SET @@autocommit = false;
            SELECT @@autocommit;
            --> 0(수동)
    
    - 로그파일에는 DML 작업만 기록된다
    - 동기화 작업 수행
      - commit을 해야만 다른 작업 공간에서도 내가 했던 작업들이 동기화가 된다

              --> capped collction이란 고정된 크기를 가진 컬렉션이다
              size 옵션을 반드시 설정해야 한다
              size가 다 찼을경우 오래된 순으로 삭제된다
          autoindexid : true로 설정하면 _id필드에 index를 자동으로 생성
          size : 해당 컬렉션의 최대 크기를 지정(byte)
          max : 컬렉션에 추가할 수 있는 최대 개수 지정

                            username : post_username,

                           {content : post_content,
